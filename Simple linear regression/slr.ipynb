```python
import pandas as pd
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_diabetes
```

```python
dia=load_diabetes()
```

```python
print(dia)
```

## ps: to predict the progression in one year after baseline provided 10 features

```python
print(dia.DESCR)
```

```python
# now we will cover this data into dataframe 
```

```python
df=pd.DataFrame(dia.data, columns=dia.feature_names)
```

```python
df
```

```python
df.head()
```

```python
# this data is scalled because all the personal information is hidden
```

```python
df['target']=dia.target
```

```python
df
```

```python
# eda, data cleaning , data pre processing , feature engineering
```

```python
x=df[['bmi']] #iv
y=df['target'] #dv
```

```python
x 
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
# random state when fixed it helps use fix the data and get us same reslt again and again
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
# model training
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model
```

```python
# training of model
model.fit(x_train,y_train)
# now at the backend it has trained the model using gardient decnet to give optimal coefficent
```

```python

model.coef_
```

```python
model.intercept_
```

```python
y_pred=model.predict(x_test)
# now we will get y_test from this and then we will compare the values with real y_test
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_test, y_test, color='black', label='Actual data')
plt.plot(x_test, y_pred, color='red', label='Predicted line', linewidth=3)
plt.xlabel("BMI")
plt.ylabel("Disease Progression")
plt.legend()
plt.show()
```

```python
plt.scatter(x_test,y_pred,color='red',label='predicated data')
```

```python
# model evaluation
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_pred)
# lower the value better the model will be
```
