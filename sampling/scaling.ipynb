```python
import kagglehub
import pandas as pd
import os

# Download the dataset from Kaggle
dataset_path = kagglehub.dataset_download("sakshisatre/tips-dataset")
print("Dataset downloaded to:", dataset_path)

# Find the CSV file inside the dataset folder
csv_files = [f for f in os.listdir(dataset_path) if f.endswith(".csv")]
if not csv_files:
    raise FileNotFoundError("No CSV files found in dataset.")

csv_path = os.path.join(dataset_path, csv_files[0])

# Load CSV into a DataFrame
df = pd.read_csv(csv_path)

# Preview DataFrame
print("First 5 rows:")
print(df.head())
```

```python
sns.displot(df['total_bill'])
```

```python
mean=np.mean(df.total_bill)
```

```python
mean
```

```python
std=np.std((df['total_bill']))
```

```python
std
```

```python
std_data=[]
for i in list(df['total_bill']):
    zscore=(i-mean)/std
    std_data.append(zscore)
```

```python
std_data
```

```python
# feature scaling doesnt change the distribution of data
sns.displot(std_data)
```
